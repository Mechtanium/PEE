## **Exploration**

### How to explore offshore

In exploring offshore, the following steps come to mind
1. **Acquire seismic data.** This is done by towing a seismic survey vessel behind a seismic source, which emits sound waves that travel through the water and into the subsurface. The seismic waves are reflected and refracted by different rock layers, and the reflected and refracted waves are detected by hydrophones towed behind the vessel. The seismic data is then processed to create an image of the subsurface geology.
2. **Identify potential drilling targets.** The seismic data is analyzed to identify areas where there is a high probability of finding oil or gas. These areas are known as prospects.
3. **Assess the risks and rewards of drilling each prospect.** This includes estimating the potential oil or gas reserves, the cost of drilling, and the environmental risks.
4. **Obtain the necessary permits and licenses from the government.**

## **Drilling**

### How to drill for oil and gas offshore
The following steps are part of the drilling process
1. **Select a platform, drilling rig and crew.**
2. **Transport the plarform to the drilling location.**
3. **Position the platform over the drilling target.**
4. **Set the conductor casing.** This is a large-diameter pipe that is driven into the seabed to protect the well from the elements.
5. **Drill the wellbore.** This is done using a rotating drill bit that is attached to a long string of drill pipe.
6. **Run wireline logs.** Wireline logs are instruments that are lowered into the wellbore to collect data about the rock formations and fluids in the well.
7. **Evaluate the well data.** The wireline log data is used to determine the presence of oil or gas, the thickness of the reservoir, and the quality of the reservoir rock.
8. **Test the well.** This is done by producing oil or gas from the well and measuring the production rate and fluid properties.
9. **Complete the well.** This involves installing the necessary production equipment, such as casing, tubing, and wellhead.

### How to select a platform
image.png
To select a platform we must know the types of platforms and how they are used.
The following are the types of offshore platforms that exist for marine offshore oil and gas drilling and production, along with the range of water depths they can be used in:

**Fixed Platforms**

* **Water depths:** 10-300 meters (33-984 feet)

Fixed platforms are the most common type of offshore platform, and are permanently attached to the seabed. They are typically used in shallow to mid-water depths.

**Compliant Towers**

* **Water depths:** 100-500 meters (328-1640 feet)

Compliant towers are a type of fixed platform that is designed to flex with the movement of the waves. This allows them to be used in deeper water depths than traditional fixed platforms.

**Semi-Submersible Platforms**

* **Water depths:** 200-3000 meters (656-9842 feet)

Semi-submersible platforms are floating platforms that are anchored to the seabed. They are typically used in deep and ultra-deep water depths.

**Jack-Up Drilling Rigs**

* **Water depths:** 10-100 meters (33-328 feet)

Jack-up drilling rigs are mobile platforms that have legs that can be extended to the seabed to support the platform. They are typically used in shallow water depths.

**Drillships**

* **Water depths:** 300-3000 meters (984-9842 feet)

Drillships are floating vessels that are used for drilling operations in deep and ultra-deep water depths.

**Floating Production Systems (FPSOs)**

* **Water depths:** 300-3000 meters (984-9842 feet)

FPSOs are floating platforms that are used to produce, store, and offload oil and gas from offshore wells. They are typically used in deep and ultra-deep water depths.

**Tension Leg Platforms (TLPs)**

* **Water depths:** 500-1500 meters (1640-4921 feet)

TLPs are floating platforms that are anchored to the seabed by tethers. This allows them to remain stable in deep water depths.

**Spar Platforms**

* **Water depths:** 1000-3000 meters (3280-9842 feet)

Spar platforms are floating platforms that have a long, cylindrical hull that is submerged in the water. This provides stability in deep water depths.

It is important to note that the water depth range for each type of offshore platform can vary depending on the specific design of the platform. Additionally, some platforms may be able to operate in deeper water depths if they are modified or upgraded.

## **Production**

### How to produce from offshore wells

Production offshore largely entails storage and transport of oil and gas. Here are some simple steps in the process.


1. **Tie the well back to a production facility.** This is done by laying a pipeline from the well to the production facility.
2. **Start production.** This involves pumping oil or gas from the well to the production facility.
3. **Monitor the well performance and make adjustments as needed.**
4. **Maintain the well and production equipment.**
5. **Transport the oil or gas to market.**
6. **Plug and abandon the well at the end of its productive life.**
7. **Remediate the drilling site and restore it to its original condition.**
